## Question 1

```python
#get a df of bathrooms in Eugene
bathroom = gdf[gdf['amenity'] == 'toilets'].reset_index()
bathroom
```

```python
#Reproject data
gdf = gdf.to_crs('EPSG:32610')
bathroom = bathroom.to_crs('EPSG:32610')
```

```python
# Get coordinates of EMU
emu = gdf[gdf['name'] == 'Erb Memorial Union (EMU)'].reset_index()
emu
```

```python
# Get bathroom and EMU centroids
bathroom['centroid'] = bathroom['geometry'].apply(
  lambda x: x.centroid if type(x) == Polygon else (
  x.centroid if type(x) == MultiPolygon else x))

emu['centroid'] = emu['geometry'].apply(
  lambda x: x.centroid if type(x) == Polygon else (
  x.centroid if type(x) == MultiPolygon else x))
```

```python
# Compute distances
emu_x = emu['centroid'].x.values[0]
emu_y = emu['centroid'].y.values[0]
distances = np.sqrt(((emu_x - bathroom['centroid'].x.values)**2)
                     + ((emu_y - bathroom['centroid'].y.values)**2))

# Add to GeoDataFrame
bathroom['euclidean_distance'] = distances
```

```python
bathroom.nsmallest(10, ['euclidean_distance'])[['name', 'euclidean_distance']]
```

```python
# Make a new DataFrame containing only the three most relevant columns
nearest_bathroom = bathroom.nsmallest(10, ['euclidean_distance'])[['name', 'euclidean_distance', 'centroid']]

# Set column geometry
nearest_bathroom = nearest_bathroom.set_geometry('centroid')

# Convert back to WGS84
nearest_bathroom = nearest_bathroom.to_crs('EPSG:4326')

# Define center of map (i.e. Condon Hall) and initial zoom level
lat_lon = [44.0451, -123.0781]
m = folium.Map(location=lat_lon, zoom_start=12)

for i in range(0, nearest_bathroom.shape[0]):
    my_string = 'name: {}, distance: {}'.format(nearest_bathroom.iloc[i]['name'], nearest_bathroom.iloc[i]['euclidean_distance'])
    folium.Marker([nearest_bathroom.iloc[i]['centroid'].y, nearest_bathroom.iloc[i]['centroid'].x],
                 popup=my_string).add_to(m)

# Display map
m
```

So, I plotted the route for my favorite coffee stand (Rogue Roasters) to overtake the Dutch Headquartes and to establish themselves as the coffee overlords.

```python
# Specify type of data
tags = {'building': True}

# Download building geometries from OSM
gpdf = ox.geometries_from_place('Grants Pass, Oregon, USA', tags)
gpdf['amenity'].unique()
```

```python
Best_Coffee = gpdf[gpdf['name'] == "Rogue Roasters Cafe"]
Dutch = gpdf[gpdf['name'] == "Dutch Bros. Headquarters"]
```

```python
# Define coordinates of GP Downtown
lat_lon = (42.439666560883545, -123.3294737289693)

# Import walkable street network data around Downtown GP
gp = ox.graph_from_point(lat_lon, dist=1600, network_type='walk')

# Convert to graph
gp_graph_proj = ox.project_graph(gp)

# Get edges and nodes separately
gp_nodes_proj, gp_edges_proj = ox.graph_to_gdfs(gp_graph_proj, nodes=True, edges=True)

# Reproject to UTM Zone 10N
gpdf = gpdf.to_crs('EPSG:32610')
Dutch = Dutch.to_crs('EPSG:32610')
Best_Coffee = Best_Coffee.to_crs('EPSG:32610')
Best_Coffee
```

```python
# Get RR and Dutch centroids
Best_Coffee['centroid'] = Best_Coffee['geometry'].apply(
  lambda x: x.centroid if type(x) == Polygon else (
  x.centroid if type(x) == MultiPolygon else x))

Dutch['centroid'] = Dutch['geometry'].apply(
  lambda x: x.centroid if type(x) == Polygon else (
  x.centroid if type(x) == MultiPolygon else x))
```

```python
#Get Rogue Roasters coordinates
CoffeeCenter = (Best_Coffee['centroid'].y.values[0], Best_Coffee['centroid'].x.values[0])
DutchCenter = (Dutch['centroid'].y.values[0], Dutch['centroid'].x.values[0])
```

```python
# Find the node in the graph that is closest to the origin point (here, we want to get the node id)
coffee_node = ox.distance.nearest_nodes(G=gp_graph_proj, X=CoffeeCenter[1], Y=CoffeeCenter[0], return_dist=False)

# Find the node in the graph that is closest to the target point (here, we want to get the node id)
dutch_node = ox.distance.nearest_nodes(G=gp_graph_proj, X=DutchCenter[1], Y=DutchCenter[0], return_dist=False)
```

```python
# Calculate the shortest path
coffee_takeover_route = nx.shortest_path(G=gp_graph_proj, source=coffee_node, target=dutch_node, weight='length')

# Plot the shortest path using folium
m = ox.plot_route_folium(gp, coffee_takeover_route, weight=5)
m
```

```python
nearest_bathroom = nearest_bathroom.to_crs('EPSG:32610')

# Get x and y coordinates of all ten of the nearest bathroom
target_xy = (nearest_bathroom['centroid'].y.values, nearest_bathroom['centroid'].x.values)
```

```python
# Define coordinates of EMU
lat_lon = (44.045397015109316, -123.07301233319605)

# Import walkable street network data around Condon Hall
g = ox.graph_from_point(lat_lon, dist=1600, network_type='walk')

# Convert to graph
graph_proj = ox.project_graph(g)

# Get x and y coordinates of EMU
orig_xy = (emu['centroid'].y.values[0], emu['centroid'].x.values[0])

# Find the node in the graph that is closest to the origin point (here, we want to get the node id)
orig_node = ox.distance.nearest_nodes(G=graph_proj, X=orig_xy[1], Y=orig_xy[0], return_dist=False)

# Get edges and nodes separately
nodes_proj, edges_proj = ox.graph_to_gdfs(graph_proj, nodes=True, edges=True)
```

```python
routes = []
distances = []
differences = []
for i in range(len(target_xy[0])):
    
    # Find the node in the graph that is closest to the target point (here, we want to get the node id)
    target_node = ox.distance.nearest_nodes(graph_proj, X=target_xy[1][i], Y=target_xy[0][i], return_dist=False)
    
    # Calculate the shortest path
    route = nx.shortest_path(G=graph_proj, source=orig_node, target=target_node, weight='length')
    
    # Append route to list
    routes.append(route)
    
    # Get the nodes along the shortest path
    route_nodes = nodes_proj.loc[route]
    
    # Create a geometry for the shortest path
    route_line = LineString(list(route_nodes['geometry'].values))

    # Create a GeoDataFrame
    route_geom = gpd.GeoDataFrame([[route_line]], geometry='geometry', crs=edges_proj.crs, columns=['geometry'])

    #calculate euclidean
    distance = np.sqrt(((emu_x - target_xy[1][i])**2)
                     + ((emu_y - target_xy[0][i])**2))
    
    # calculate difference
    diff = (route_geom['geometry'].length[0] / 1000) - (distance / 1000)
    
    # Print length of route
    print('Walking distance is:' + str(route_geom['geometry'].length[0] / 1000) + ' km.\nEuclidean distance is: ' + str(distance / 1000) + " km.\n\tDifference is: " + str(diff) + " km.")
    
    # Append distances to list
    distances.append(route_geom['geometry'].length[0])
    
    #Append differences to list
    differences.append(abs(diff))
    
print("\nThe average absolute difference between Euclidean and walking distances is " + str(sum(differences)/10) + " km.")
```

### 3b)

Euclidean distances, colloquially the distance 'as the crow flies' are a direct route from one object to another. This may be helpful for intense computational calculations where it is necessary to simplify your measurements in order to speed up processing times. It also may be helpful for instances where other factors/barriers may not make much of a difference in travel time (such as steam boat travel across the Pacific Ocean). On the other hand, it is not very helpful for understanding the networks between places. If data needs a finer level of precision (such as travel times at a minute data scale), or if there are substantial barriers to traveling from one point to another, then using a network analysis is more poignant.
